```python
import numpy as np
from scipy import stats
```

```python
ptratio = boston["PTRATIO"]
t_stat, p_value = stats.ttest_1samp(ptratio, 18)
print("Null Hypothesis (H0): The mean pupil-to-teacher ratio = 18")
print("Alternative Hypothesis (Ha): The mean pupil-to-teacher ratio  18")
print(f"\nTest Statistic (t): {t_stat:.3f}")
print(f"P-value: {p_value:.4f}")
```

```python
alpha = 0.05
if p_value < alpha:
    print("\nConclusion: Reject the null hypothesis  the mean PTRATIO is significantly different from 18.")
else:
    print("\nConclusion: Fail to reject the null hypothesis  no significant difference from 18.")
```

```python
import statsmodels.api as sm
```

```python
X = boston[['DIS']]
y = boston['MEDV']    
```

```python
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
```

```python
print(model.summary())

coef_DIS = model.params['DIS']
print(f"\nCoefficient for DIS: {coef_DIS:.4f}")
```

```python
if coef_DIS > 0:
    print(f"Interpretation: For every 1 unit increase in weighted distance (DIS), "
          f"the median home value (MEDV) increases by approximately {coef_DIS:.2f} units, "
          f"holding other factors constant.")
else:
    print(f"Interpretation: For every 1 unit increase in weighted distance (DIS), "
          f"the median home value (MEDV) decreases by approximately {abs(coef_DIS):.2f} units, "
          f"holding other factors constant.")
```
